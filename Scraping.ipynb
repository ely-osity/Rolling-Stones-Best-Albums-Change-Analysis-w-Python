```python
import csv
import pandas as pd

def dataCollect(url, page, fileName):
    files = {}
    URL = url+'?page=1'
    tablelist = pd.read_html(URL)[0]
    #csv合并
    for i in range(2,page+1):#自动翻页
        URL = url+'?page={}'.format(i)
        tableatc = pd.read_html(URL)[0]
        tablelist = pd.concat([tablelist, tableatc])
    tablelist.to_csv(fileName, index = False)#存入文件
    return tablelist
```
